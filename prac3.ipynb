```python
import pandas as pd

# Load Titanic dataset from seaborn
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
summary_stats = titanic.groupby('pclass')['age'].describe()
summary_stats
```

```python
mean_age = titanic.groupby('pclass')['age'].mean()
median_age = titanic.groupby('pclass')['age'].median()
min_age = titanic.groupby('pclass')['age'].min()
max_age = titanic.groupby('pclass')['age'].max()
std_age = titanic.groupby('pclass')['age'].std()
```

```python

print("\nIndividual Summary Statistics for Age grouped by Pclass:")
print("Mean Age:", mean_age)
print("Median Age:", median_age)
print("Min Age:", min_age)
print("Max Age:", max_age)
print("Standard Deviation:", std_age)
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
```

```python
df.head()
```

```python
df['species']=pd.Categorical.from_codes(iris.target,iris.target_names)
```

```python
df.head()
```

```python
species=df['species'].unique()
```

```python
for sp in species:
    print('Describing ',sp)
    d=df[df['species']==sp]
    print(d.describe())
```
